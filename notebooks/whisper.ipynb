```python
from faster_whisper import WhisperModel
```

```python
model = WhisperModel(
    "medium.en", 
    device="cuda", 
    compute_type="float16"
)
```

```python
from pathlib import Path

import numpy as np

def load_pcm(path: Path):
    audio = np.fromfile(path, dtype=np.int16)
    return audio.astype("float32") / 32768.0
```

```python
audio = load_pcm(Path("/home/andrew/Documents/audio/server/.cache/8f29b2c2-424b-41ec-a5a8-ceb38b550321-1.pcm"))[:44000]
```

```python
segments, info = model.transcribe(audio, beam_size=1, word_timestamps=False, vad_filter=False)
```

```python
for segment in segments:
    print(f"[{segment.start:.2f}s -> {segment.end:.2f}s] {segment.text}")
    print(segment)
```
